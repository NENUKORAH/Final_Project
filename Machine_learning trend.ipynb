```python
plt.figure(figsize=(15,8))
sns.lineplot(data=tesla.iloc[:,:-2])
plt.ylabel("Stock Value")
plt.title("Stock trend")
```

```python
def get_sales_from_date(date1="2011-01-01",date2="2021-09-10",window=50):
    return(tesla.loc[f"{date1}":f"{date2}"].rolling(window).mean())
```

```python
def visualize_from_dates(df):
        return(plt.figure(figsize=(15,8)),
           sns.lineplot(data=df.iloc[:,4:5]))
```

```python
visualize_from_dates(get_sales_from_date("2010-01-01","2021-09-10",10))
plt.title("Adjusted close trend 2010 - 2021")
```
